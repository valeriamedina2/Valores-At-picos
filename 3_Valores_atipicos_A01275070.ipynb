#ventas_precios_corrientes

```python
fig = plt.figure(figsize =(5, 3))
plt.hist(x=df["ventas_precios_corrientes"], color='lightblue', rwidth=0.50)
plt.title('Histograma ventas precio corrientes ')
plt.xlabel('Venta Precio Corriente')
plt.ylabel('Frecuencia')

plt.show() #dibujamos el histograma
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(10, 6))
plt.boxplot(df["ventas_precios_corrientes"]) 
plt.title("Outliers de Precio Corriente")
plt.show() #dibujamos el diagrama
```

##Metodo dde desviacion estandar 


```python
y=df["ventas_precios_corrientes"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers= df[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=df["ventas_precios_corrientes"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= df[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr1= df[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr1
```

```python
#Realizamos diagrama de caja o bigote
fig1 = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr1["ventas_precios_corrientes"]) 
plt.title("Outliers ventas precio corrientes")
plt.show() #dibujamos el diagrama
```

```python
data_clean_iqr1.to_csv("ventas_precio_corrientes.csv")
from google.colab import files
files.download("ventas_precio_corrientes.csv")
```
